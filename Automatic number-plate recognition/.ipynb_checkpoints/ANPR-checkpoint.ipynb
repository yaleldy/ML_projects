# Import Libraries

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import os
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import numpy as np
import easyocr
```

# plot the number plot

```python
plate_path = 'Sample_Images/Datacluster_number_plates (96).jpg'
img = cv2.imread(plate_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
```

```python
blur_img = cv2.GaussianBlur(gray,(3,3),cv2.BORDER_DEFAULT)
edge = cv2.Canny(blur_img, 0,0)
plt.imshow(cv2.cvtColor(edge, cv2.COLOR_BGR2RGB))
```

# Contours and mask

```python
import imutils
keypoints = cv2.findContours(edge.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(keypoints)
contours = sorted(contours, key=cv2.contourArea, reverse=True)[:10]
location = None
for contour in contours:
    approx = cv2.approxPolyDP(contour, 10, True)
    if len(approx) == 4:
        location = approx
        break
```

```python
location
```

```python
mask = np.zeros(gray.shape, np.uint8)
new_image = cv2.drawContours(mask, [location], 0,255, -1)
new_image = cv2.bitwise_and(img, img, mask=mask)
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
```

```python
crop_img = img[location[0][0][1]:location[2][0][1],location[0][0][0]:location[2][0][0]]
```

```python
plt.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(crop_img)
result
```
