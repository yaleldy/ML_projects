# Import Libraries

```python
import tensorflow as tf
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from tensorflow.keras import datasets, layers, models
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import TensorBoard
import os
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import numpy as np
import mediapipe as mp
```

# Use Mediapipe

```python
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands
```

```python
def hands_process(image, hands):
    image.flags.writeable = False
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(image)    
    # Draw the hand annotations on the image.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    return image, results
```

```python
def drawing_landmarks (image, hands):
    if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
        mp_drawing.draw_landmarks(
            image,
            hand_landmarks,
            mp_hands.HAND_CONNECTIONS,
            mp_drawing.DrawingSpec(
            color=(255,0,0),
            thickness=3,
            circle_radius=3
          ),
          mp_drawing.DrawingSpec(
            color=(0,0,255),
            thickness=2,
            circle_radius=2
          ))
```

```python
cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      print("Ignoring empty camera frame.")
      # If loading a video, use 'break' instead of 'continue'.
      continue

    image, results = hands_process(image, hands)
    drawing_landmarks (image, results)
    
    cv2.imshow('MediaPipe Hands', image)
    if cv2.waitKey(5) & 0xFF == ord('q'):
      break

cap.release()
cv2.destroyAllWindows()
## Remove this would freeze the macbook for some reason
cv2.waitKey(1)
```

```python
image_cvt = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_cvt)
```

```python
## take a look at the landmark data
def landmark_values (results):
    landmarks = []
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
                for landmark_i in hand_landmarks.landmark:
                    temp = np.array([landmark_i.x, landmark_i.y, landmark_i.z])
                    landmarks.append(temp)
    return np.array(landmarks)                
```

# Capture training and test data

### create empty folder

```python
n_capture = 40
n_class = 5
sign_class = ['one', 'two', 'three', 'four', 'five']

for i in range(0,n_class):
    for j in range(0, n_capture):
        os.makedirs(sign_class[i] + "/" + "{num_capture}".format(num_capture = j),exist_ok=True)
```

### create training and test data

```python
from datetime import datetime
duration = 2
cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
    for i in range(0, n_class):
        for j in range (0, n_capture):
            start_time = datetime.now()
            diff = (datetime.now() - start_time).seconds 
            while( diff <= duration ):
                success, image = cap.read()
                image, results = hands_process(image, hands)
                drawing_landmarks (image, results)
                cv2.putText(image, "New Capture starts in: "+str(duration - diff), (70,70), cv2.FONT_HERSHEY_SIMPLEX , 1, (255, 0, 0), 2, cv2.LINE_AA)# adding timer text

                cv2.putText(image,
                'collecting: sign {sign_class} capture number {num_capture}'.format(sign_class=sign_class[i], num_capture = j), 
                (30,30), cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1,1)

                cv2.imshow('MediaPipe Hands', image)

                diff = (datetime.now() - start_time).seconds
                
                if diff == 2:
                    training_data = landmark_values(results)
                    np.save(sign_class[i] + "/" + "{num_capture}".format(num_capture = j) + "/landmarks", training_data)


                if cv2.waitKey(5) & 0xFF == ord('q'):
                  break
            else:
                continue

            break
        
        else:
            continue

        break

    cap.release()
    cv2.destroyAllWindows()
    ## Remove this would freeze the macbook for some reason
    cv2.waitKey(1)
```

```python
## load in the collected data
x_sign = []
y_sign = []
for i in range(0, n_class):
    for j in range (0, n_capture):
        datai = np.load(sign_class[i] + "/" + "{num_capture}".format(num_capture = j) + "/landmarks.npy")
        x_sign.append(datai)
        y_sign.append(i)

x_sign = np.array(x_sign)
y_sign = np.array(y_sign)
y_sign_cat = tf.keras.utils.to_categorical(y_sign, dtype='int')
```

```python
x_sign.shape
y_sign_cat.shape
```

```python
## separate training and test data
x_train, x_test, y_train, y_test = train_test_split(x_sign, y_sign_cat, test_size=0.2)
```

# Build CNN model

```python
model = models.Sequential()
model.add(layers.LSTM(64, return_sequences=True, activation='relu', input_shape=(21,3)))
model.add(layers.LSTM(128, return_sequences=True, activation='relu'))
model.add(layers.LSTM(64, return_sequences=False, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(5, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs = 1000, validation_data = (x_test, y_test))
```

```python
# accuracy plot
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
```

```python
model.save('weights.h5')
```

# Validate the model and plot confusion matrix

```python
# Validate test data
y_test_pred = model.predict(x_test)
y_test_pred_num = np.argmax(y_test_pred, axis = 1)
y_label = np.argmax(y_test, axis = 1)
```

```python
# confusion matrix
from sklearn.metrics import multilabel_confusion_matrix, accuracy_score
multilabel_confusion_matrix(y_label, y_test_pred_num)
```

```python
test_loss, test_acc = model.evaluate(x_test,  y_test)
```

# Validate in real time video

```python
cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      print("Ignoring empty camera frame.")
      # If loading a video, use 'break' instead of 'continue'.
      continue

    image, results = hands_process(image, hands)
    drawing_landmarks (image, results)
    landmarks = landmark_values(results)
    if landmarks.shape == (21,3):
        y_pred = model.predict(np.expand_dims(landmarks, axis=0))[0]
        if np.max(y_pred) > 0.6:
            cv2.putText(image,
                        'predicted sign: '+ str(np.argmax(y_pred)+1), 
                        (30,30), cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),2,2)
            cv2.putText(image, 'predicted percent: ' + str(np.round(np.max(y_pred),2)),
                       (30,60), cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2,2)

    
    cv2.imshow('MediaPipe Hands', image)
        
    if cv2.waitKey(5) & 0xFF == ord('q'):
      break

cap.release()
cv2.destroyAllWindows()
## Remove this would freeze the macbook for some reason
cv2.waitKey(1)
```
